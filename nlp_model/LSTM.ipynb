```python
import pandas as pd
import nltk
from sklearn.model_selection import train_test_split
from pre_process import PreProcess
```

```python
import pandas as pd

data = pd.read_csv('./spam.csv')
print(data)
```

```python
print(type(data["Message"]))
data["Message"] = PreProcess.preProcess(data["Message"])
print(data["Message"][0])
```

```python
print(data["Message"][0])
print(data["Message"].size)
for i in range(0, data["Message"].size):
  data["Message"][i] = PreProcess.standardize(data["Message"][i])
print(data["Message"][0])
```

```python
data["Label"] = 0
for i in range(0, data["Category"].size):
  data["Label"][i] = PreProcess.transformLabel(data["Category"][i])
```

```python
X = data["Message"]
Y = data["Label"]
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size=0.8, test_size=0.2, random_state=0)
print(X_train, X_test, Y_train, Y_test)
```

```python
from gensim.models import Word2Vec
word_to_vec_model = Word2Vec(X_train, vector_size=150, window=5, min_count=1, workers=4)
word_to_vec_model.save("word2vec.model")
```

```python
print(word_to_vec_model.wv[0])
test = word_to_vec_model.wv.most_similar('ok', topn=10)
print(test)
```
